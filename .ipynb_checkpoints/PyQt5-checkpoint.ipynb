```python
class MyWindow(QMainWindow):
    def __init__(self):
        super(MyWindow, self).__init__()
        self.setGeometry(200,200,450,300)
        self.setWindowTitle('Guess The Word')
        self.initUI()
        threading.Thread(target=self.simulate,daemon=True).start()
        #self.simulate()
    
    def initUI(self):
        self.chances = QtWidgets.QLabel(self)
        self.chances.setText("remaining chances - 100")
        self.chances.resize(150,50)
        self.chances.move(50,50)
        
        self.questWord = QtWidgets.QLabel(self)
        self.questWord.setText("_ _ _ _ _ _ _ _")
        self.questWord.move(50,120)
        
#         self.b1 = QtWidgets.QPushButton(self)
#         self.b1.setText('Click')
#         self.b1.clicked.connect(self.onClick)
        
        self.txtbox = QtWidgets.QLineEdit(self)
        self.txtbox.move(280,150)
        self.txtbox.resize(50,50)
        self.txtbox.setAlignment(Qt.AlignCenter)
        validator = QRegExpValidator(QRegExp("[A-Za-z]{1}"))
        self.txtbox.setValidator(validator)
        self.txtbox.setPlaceholderText('enter a letter')
        
#     def onClick(self):
#         self.label.setText('clicked')
#         self.update()
    
#     def update(self):
#         self.label.adjustSize()
    
    def getWord(self, level=0):
        self.Word='DISTRCI'
    
    def generate(self):
        L= len(self.Word)
        self.filWord = [False]* L
        c= int(0.3*L)
        if c==0: 
            c+=1
        while c>0:
            i = random.randint(0, L-1)
            ch=self.Word[i]
            ind = -1
            while(True):
                ind = self.Word.find(ch, ind+1)
                if(ind==-1):
                    break
                self.filWord[ind] = True
                c=c-1
    
    def inp(self):
        while True:
            alpha = self.txtbox.text()
            if alpha!='' and alpha.isalpha():
                self.inputch=alpha.upper()
                return 
            else:
                continue

    def displayW(self):
        i=0
        str=''
        for c in self.Word:
            if(self.filWord[i]):
                str+=c
            else:
                str+='_'
            i+=1
        self.questWord.setText(str)
    
    def checks(self,c):
        p=self.Word.find(c)
        if (p==-1):
            return -1

        if(self.filWord[p]):
            return 0

        ind = -1
        while(True):
            ind = self.Word.find(c, ind+1)
            if ind==-1:
                break
            self.filWord[ind] = True
        return 1
    
    def simulate(self):
        self.getWord()
        self.generate()
        chance = 7
        while( (chance!=0) and (False in self.filWord) ):
            self.txtbox.setText('')
            self.chances.setText("Chances Remaining - "+str(chance))
            self.displayW()
            x=threading.Thread(target=self.inp,daemon=True)
            x.start()
            x.join()
            ch = self.inputch
            s = self.checks(ch)
            if s==-1:
                print('Wrong Guess!')
                chance-=1
                continue
            if s==0:
                print('Try New Alphabet')
                continue
            if s==1:
                print('voila! nice guess')
                continue

        if(not(False in self.filWord)):
            print('Amazing, u have cracked the word')
        else:
            print('Chances perished! better luck next time ')
               

def window():
    app = QApplication(sys.argv)
    win = MyWindow()
    win.show()
    sys.exit(app.exec_())
```

```python
## get a word by length filter
def getWord(level=0):
    word='DISTRCI'
    return word
```

```python
## display 30% of the letters, avoid first and last letter together
def generate(word):
    L= len(word)
    dispWord = [False]* L
    c= int(0.3*L)
    if c==0: 
        c+=1
    while c>0:
        i = random.randint(0, L-1)
        ch=word[i]
        ind = -1
        while(True):
            ind = word.find(ch, ind+1)
            if(ind==-1):
                break
            dispWord[ind] = True
            c=c-1
            
    return dispWord
    
```

```python
## accept only alphabet input
def inp():
    print('->',end='')
    while True:
        alpha = str(input())[0]
        if alpha.isalpha():
            return alpha.upper()
        else:
            continue
```

```python
## check if alphabet is present,, update remaining chances and display word
def checks(word, wdis,c):
    p=word.find(c)
    if (p==-1):
        return -1
    
    if(wdis[p]):
        return 0
    
    ind = -1
    while(True):
        ind = word.find(c, ind+1)
        if ind==-1:
            break
        wdis[ind] = True
    return 1
```

```python
def display(wrd, diwrd):
    i=0
    for c in wrd:
        if(diwrd[i]):
            print(c, end=' ')
        else:
            print('_', end=' ')
        i+=1
    print()  
```

```python
def simulate():
    wrd = getWord()
    diwrd = generate(wrd)
    chances = 7
    while( (chances!=0) and (False in diwrd) ):
        print('Chances Remaining - ', chances)
        display(wrd, diwrd)
        ch = inp()
        s = checks(wrd, diwrd, ch)
        if s==-1:
            print('Wrong Guess!')
            chances-=1
            continue
        if s==0:
            print('Try New Alphabet')
            continue
        if s==1:
            print('voila! nice guess')
            continue
    
    if(not(False in diwrd)):
        print('Amazing, u have cracked the word')
    else:
        print('Chances perished! better luck next time ')
```

```python
simulate()
```
