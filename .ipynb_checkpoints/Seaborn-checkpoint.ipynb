```python
import seaborn as sns
import pandas as pd
```

```python
sales= pd.read_excel('sales.xls')
sales.head()
```

```python
sns.relplot('Sales','Profit', data=sales, hue='Order Priority', style='Ship Mode', size='Discount', sizes=(50,200))
```

```python
sales['Year']=sales['Order Date'].dt.year
print(sales.Year.unique())
sns.relplot('Year', 'Sales', data=sales, kind='line')
```

```python
sns.relplot('Year', 'Sales', data=sales, kind='line', estimator=None)
```

```python
sns.relplot('Year', 'Sales', data=sales, kind='line', estimator=None,hue='Ship Mode', col='Region', col_wrap=3)
```

```python
sns.catplot('Ship Mode', 'Sales', data=sales.query("Profit > 100"))
```

```python
sns.catplot('Ship Mode', 'Sales', data=sales.query("Profit > 100"), kind='swarm')
```

```python
sns.catplot('Product Category', 'Shipping Cost', data=sales, kind='box')
```

```python
sns.catplot('Product Category', 'Shipping Cost', data=sales, kind='violin')
```

```python
sns.catplot('Product Category', 'Shipping Cost', data=sales, kind='bar')
```

```python
sns.catplot('Product Category', 'Shipping Cost', data=sales, kind='point')
```

```python
sns.distplot(sales.Sales)
```

```python
sns.distplot(sales.Profit, hist=False)    ##kde=False
```

```python
sns.jointplot("Discount","Product Base Margin", data=sales, kind='hex')
```

```python
sns.jointplot("Discount","Product Base Margin", data=sales, kind='kde')
```

```python
sns.pairplot(sales)
```
