```python
titanic= pd.read_csv("titanic.csv")
print(titanic.head(5))
print(titanic.tail(5))
```

```python
subdf=titanic[["Age","Sex"]]
print(type(subdf))
print(subdf.shape)
print(subdf)
print(titanic[titanic["Age"]>70])
print(titanic[(titanic["Pclass"]==2) | (titanic["Pclass"]==3)])
print(subdf["Age"].notna())
print(subdf[subdf["Age"].notna()])
```

```python
print(subdf.loc[subdf["Age"]>35, ["Sex","Age"]])
print(subdf.iloc[2:5,:1])  ##print(subdf.loc[2:5,"Age"])
```

```python
subdf["look"]=(subdf["Age"]/100*2)
print(subdf.head())
```

```python
subdfRename=subdf.rename( columns={"look":"looks"})
subdfRename.head()
```

```python
subdfRename.rename(columns=str.lower, inplace=True) 
subdfRename.head()
```

```python
titanic.agg({
    "Age":["max","mean","median","skew"],
    "Fare":["max","mean","median","skew"]
})
```

```python
for f in titanic[["Age", "Sex"]].groupby("Sex"):
    print(f)
titanic.groupby("Sex")["Age"].describe()    ##titanic[["Age", "Sex"]].groupby("Sex").describe()
```

```python
for f in titanic.groupby(["Sex", "Embarked"])["Age"]:
    print( f)
titanic.groupby(["Sex", "Embarked"])["Age","Pclass"].mean()
```

```python
titanic["Sex"].value_counts()   ##titanic.groupby("Sex")["Sex"].count()
```

Feature Engineering

```python
heights= pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/FeatureEngineering/2_outliers_z_score/heights.csv')
```

```python
uplim= heights['height'].quantile(0.99)
lolim= heights['height'].quantile(0.01)
print('outliers - ')
heights[(heights['height']>uplim) | (heights['height']<lolim)]
```
